```python
%load_ext autoreload
%aimport metropolis
%autoreload 1
from metropolis import *

from itertools import count
import numpy as np
from collections import namedtuple
from tqdm import tnrange, tqdm_notebook

LATTICE_SIZE = 10
# N_CONFIGS_PER_SAMPLE = 10
N_STEPS_STABILIZE = -1
STABLE_STD_E = 0.1
# N_STEPS_AVG = 1000


config = create_random_config(LATTICE_SIZE)
T = 0.1

e = []
m = []
for i in count():
    step(config, START_T)  
    e.append(calcEnergy(config))
    m.append(mag(config))
    if i > 200 and np.std(m[-200:]) <= STABLE_STD_E:
        N_STEPS_STABILIZE = i
        break
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
print("T: %s" % T)
plt.plot(e)
plt.ylabel('Energy')
plt.xlabel('Step')
plt.show()

plt.plot(m)
plt.ylabel('Mag')
plt.xlabel('Step')
plt.show()
```
